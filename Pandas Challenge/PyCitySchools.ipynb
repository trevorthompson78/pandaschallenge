```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])


school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
Total_number_of_schools = len(school_data_complete["School ID"].unique())
Total_number_of_students = len(school_data_complete["student_name"].unique())
Total_budget = (school_data_complete)["budget"].sum
Average_math_score = (school_data_complete)["math_score"].mean()
Average_reading_score = (school_data_complete)["reading_score"].mean()
Average_passing_rate = (Average_math_score + Average_reading_score)/2
school_data_complete["passing math"] = school_data_complete["math_score"]>=70
Percentage_students_passing_math = ((school_data_complete["passing math"]).mean())*100
school_data_complete["passing reading"] = school_data_complete["reading_score"]>=70
Percentage_students_passing_reading = ((school_data_complete["passing reading"]).mean())*100
Overall_Passing_Rate = (Percentage_students_passing_math + Percentage_students_passing_reading)/2
# Total_number_of_schools
# Total_number_of_students
# Total_budget
# Average_math_score
# Average_reading_score
# Average_passing_rate
# Combined_scores
# Percentage_students_passing_math
# Percentage_students_passing_reading
# Overall_Passing_Rate

Student_Grades = pd.DataFrame({"Total Schools":Total_number_of_schools, "Number of Students":Total_number_of_students, "Total Budget":Total_budget, "Average Math Score":Average_math_score, "Average Reading Score":Average_reading_score, "% Passing Math":Percentage_students_passing_math, "% Passing Reading":Percentage_students_passing_reading, "% Overall Passing":Overall_Passing_Rate}, index=[0])
Student_Grades
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_data_complete["passing math"] = school_data_complete["math_score"]>=70
school_data_complete["passing reading"]= school_data_complete["reading_score"]>=70
school_data_complete

Top_Performers = school_data_complete.groupby(["school_name"]).mean()
Top_Performers["Per Student Budget"] = Top_Performers["budget"]/Top_Performers["size"]

Top_Performers["% Passing Reading"] = round(Top_Performers["passing math"]*100,2)

Top_Performers["% Passing Math"] = round(Top_Performers["passing reading"]*100,2)
Top_Performers["% Overall Passing"] = (Top_Performers["% Passing Reading"] + Top_Performers["% Passing Math"])/2

# Top_Performers

Top_Performers_Summary = pd.merge(Top_Performers, school_data)
Top_Performers_Summary

# del Top_Performers_Summary["size"]
del Top_Performers_Summary["reading_score"]
del Top_Performers_Summary["math_score"]
del Top_Performers_Summary["Student ID"]
del Top_Performers_Summary["School ID"]
del Top_Performers_Summary["passing math"]
Top_Performers_Summary.rename(columns={"size":"Number of Students", "school_name":"School", "type":"Type"}, inplace=True)

Top_Performers_Summary

Top_Performers_Summary.sort_values(by=["% Overall Passing"], ascending=False, inplace=True)
Top_Performers_Summary.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
school_data_complete["passing math"] = school_data_complete["math_score"]>=70
school_data_complete["passing reading"]= school_data_complete["reading_score"]>=70
school_data_complete

Bottom_Performers = school_data_complete.groupby(["school_name"]).mean()
Bottom_Performers["Per Student Budget"] = Top_Performers["budget"]/Top_Performers["size"]

Bottom_Performers["% Passing Reading"] = round(Top_Performers["passing math"]*100,2)

Bottom_Performers["% Passing Math"] = round(Top_Performers["passing reading"]*100,2)
Bottom_Performers["% Overall Passing"] = (Top_Performers["% Passing Reading"] + Top_Performers["% Passing Math"])/2

# Top_Performers

Bottom_Performers_Summary = pd.merge(Bottom_Performers, school_data)
Bottom_Performers_Summary

# del Top_Performers_Summary["size"]
del Bottom_Performers_Summary["reading_score"]
del Bottom_Performers_Summary["math_score"]
del Bottom_Performers_Summary["Student ID"]
del Bottom_Performers_Summary["School ID"]
del Bottom_Performers_Summary["passing math"]
Bottom_Performers_Summary.rename(columns={"size":"Number of Students", "school_name":"School", "type":"Type"}, inplace=True)

Bottom_Performers_Summary

Bottom_Performers_Summary.sort_values(by=["% Overall Passing"], ascending=True, inplace=True)
Bottom_Performers_Summary.head()
```
